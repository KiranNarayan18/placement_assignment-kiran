````python
import pandas as pd
df = pd.read_csv('question4_solution.csv')
df
````

````python
# 7.1
from datetime import datetime
# Get all the Earth meteorites that fell before the year 2000

def get_year(datetime_str:str):

    if not isinstance(datetime_str,str):
        return None
 
    datetime_obj = datetime.strptime(datetime_str, "%Y-%m-%dT%H:%M:%S.%f")
    year = datetime_obj.year

    return year

# df['only_year']  = df['year'].apply(lambda i: get_year(i) )
# df.drop(columns=['only_year'], inplace=True)
df[df['year'].apply(lambda i: get_year(i) ) < 2000]['name']
````

````python
# 7.2
# Get all the earth meteorites co-ordinates who fell before the year 1970
df[df['year'].apply(lambda i: get_year(i) ) < 1970]['geolocation.coordinates']
````

````python
#7.3
# Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more
# than 10000kg

df[df['mass'] > 10000]['name']
````
