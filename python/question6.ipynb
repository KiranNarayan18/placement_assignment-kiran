````python
# 6.1

import pandas as pd

df = pd.read_excel('question3_solution.xlsx')

#printing all the pokeman names whose spawn rate is less than 5%
#spawn rate less than 5%
df[df['avg_spawns'] < 5]['name'] 
````

````python
#6.2 

#the elements in the column weakness are string type it should be converted back to list to find the proper length
#either we can split the list or use eval function 
#then the length of each list is calculated

df[df['weaknesses'].apply(lambda i: len(eval(i))) < 4]['name']
````

````python
# 6.3

#check for the unique values there are no empty lists
#therefore if there is no multipliers then it should 'nan'

#printing all the Pokemons that have no multipliers at all
df[df['multipliers'].isnull()]['name']

````

````python
# 6.4
# Get all Pokemons that do not have more than 2 evolutions
#assuming two evolutions are next evolutions
import json
df['num_of_evolution'] = df['next_evolution'].apply(lambda i: i.replace("'",'"') if type(i) != float else i)
df['num_of_evolution'] = df['num_of_evolution'].apply(lambda i: json.loads(i) if type(i) != float else []) 

df[df['num_of_evolution'].apply(len) > 2]
#check with the unique values of this particular column we can notice only one pokemon has more than two next evolutions
````

````python
# 6.5

from datetime import datetime, timedelta
import numpy as np

def convert_time(time_str):
    #there are some null values which can't be converted into timestamp
    #returning None
    if not isinstance(time_str,str):
        return None

    #by crosschecking the unique value for the column 
    #have noticed there are no hours , i.e., all the elements is in minutes

    #and also as the column description suggests the as minutes:seconds

    minutes, seconds = map(int, time_str.split(':'))
    #converting str to timedelta
    time_obj = datetime(1, 1, 1, 1, minute=minutes, second=seconds)

    total_seconds = timedelta(minutes=time_obj.minute, seconds=time_obj.second).total_seconds()
    
    return total_seconds

df[df['spawn_time'].apply(lambda i: convert_time(i)) < 300][['name', 'spawn_time']]
````

````python
#6.6

#all Pokemon who have more than two types of capabilities
df[df['type'].apply(lambda i: len(i)) > 2 ][['name','type']]
````
